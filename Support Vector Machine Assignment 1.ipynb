### Q1. What is the mathematical formula for a linear SVM?


Given a training dataset with n samples and m features, the input data can be represented as X = {x1, x2, ..., xn}, where xi is a m-dimensional feature vector for each sample. The corresponding class labels for each sample are y = {y1, y2, ..., yn}, where yi is the class label for the sample xi.

The goal of a linear SVM is to find a hyperplane that best separates the data into different classes. The hyperplane can be represented by the equation:

f(x) = w · x + b

where:

w is the weight vector that is perpendicular to the hyperplane.
x is the input feature vector.
b is the bias term.
The decision function is then used to classify new data points:

predicted class = sign(f(x))

The linear SVM aims to find the optimal hyperplane that maximizes the margin between the two classes. The margin is the distance between the hyperplane and the closest data points from each class. The optimization problem for finding the hyperplane can be formulated as follows:

Maximize:
2 / ||w||

subject to:
y_i(w · x_i + b) ≥ 1, for all i = 1, 2, ..., n

where:

yi is the class label for the sample xi.
w · x_i is the dot product between the weight vector and the feature vector of sample xi.
The optimization problem is typically solved using various numerical optimization techniques, such as the Sequential Minimal Optimization (SMO) algorithm or gradient descent.

After finding the optimal weight vector w and bias term b, the decision boundary is determined by the hyperplane defined by f(x) = w · x + b, and new data points can be classified based on their positions relative to the hyperplane.


###  Q2. What is the objective function of a linear SVM?


Maximize:
  2 / ||w||

subject to the constraints:
  y_i(w · x_i + b) >= 1 for all i = 1, 2, ..., n

where:
  w is the weight vector that is perpendicular to the hyperplane.
  x_i is the feature vector of the i-th training sample.
  b is the bias term of the hyperplane.
  y_i is the class label for the i-th training sample, which can be either +1 or -1 depending on the class it belongs to.
  n is the number of training samples.


The term `2 / ||w||` represents the margin. Maximizing `2 / ||w||` ensures that the hyperplane is as far away from the data points of both classes as possible, which promotes better generalization to new, unseen data.

The constraints `y_i(w · x_i + b) >= 1` ensure that all the training samples are correctly classified and lie on the correct side of the margin, with a minimum distance of 1. These constraints help in finding a robust decision boundary that best separates the two classes.

The goal of solving the optimization problem is to find the optimal weight vector `w` and bias term `b` that satisfy the constraints while maximizing the margin `2 / ||w||`. Once the optimization is performed, the decision boundary can be determined using `f(x) = w · x + b`, and new data points can be classified based on their positions relative to the hyperplane.

### Q3. What is the kernel trick in SVM?

The kernel trick in SVM is a method used to transform data into a higher-dimensional space without explicitly computing the transformation. It allows SVM to efficiently handle non-linearly separable data by finding a hyperplane in a higher-dimensional feature space, even when the data points are not linearly separable in the original input space. The kernel trick avoids the computational burden of explicitly mapping the data to the higher-dimensional space, making SVM computationally efficient.

In the kernel trick, instead of operating directly in the original feature space, the SVM operates in a feature space implicitly defined by a kernel function. The kernel function is a similarity function that calculates the inner product between two points in the higher-dimensional feature space, based on their positions in the original input space. Commonly used kernel functions include the linear kernel, polynomial kernel, radial basis function (RBF) kernel, and sigmoid kernel.

By using the kernel trick, the SVM can implicitly find a hyperplane that corresponds to a non-linear decision boundary in the original input space. This is achieved by efficiently computing the dot products between data points in the higher-dimensional space without explicitly transforming the data.

In summary, the kernel trick is a powerful technique that enables SVM to handle non-linearly separable data by implicitly mapping it to a higher-dimensional space, where the data might become linearly separable. It allows SVM to leverage the advantages of linear classifiers while dealing with complex and non-linear decision boundaries.

### Q4. What is the role of support vectors in SVM Explain with example

 Support vectors are the data points that are closest to the hyperplane in a support vector machine (SVM). They play a critical role in SVMs, as they determine the position and orientation of the hyperplane. The goal of SVM is to maximize the margin between the two classes, and the support vectors are the ones that define this margin.

To illustrate the role of support vectors, let's consider a simple example of classifying two types of flowers, roses and tulips. We have a training set of data points, each of which is labeled as either a rose or a tulip. The SVM algorithm will find a hyperplane that separates the two classes of data points. The support vectors are the data points that are closest to the hyperplane.

In the following diagram, the blue points represent roses and the red points represent tulips. The black line is the hyperplane that separates the two classes. The support vectors are the blue and red points that are closest to the hyperplane.



As you can see, the support vectors are the ones that define the margin between the two classes. The margin is the distance between the hyperplane and the closest data points. The larger the margin, the more confident the SVM is in its predictions.

In general, the more support vectors an SVM has, the more accurate it will be. However, the number of support vectors can also be a trade-off with computational resources. SVMs with a large number of support vectors can be computationally expensive to train.


### Q5. Illustrate with examples and graphs of Hyperplane, Marginal plane, Soft margin and Hard margin in SVM?


**Hyperplane**

A hyperplane is a dividing line that separates two classes of data points. In SVM, the hyperplane is chosen to maximize the margin between the two classes. The margin is the distance between the hyperplane and the closest data points.




**Marginal plane**

The marginal plane is a parallel hyperplane that is closest to the support vectors. The margin is the distance between the hyperplane and the marginal plane.



**Soft margin**

In soft margin SVM, some data points are allowed to be misclassified. This is done by relaxing the constraint of maximizing the margin. The goal of soft margin SVM is to minimize the number of misclassified data points while still maximizing the margin.

**Hard margin**

In hard margin SVM, no data points are allowed to be misclassified. This means that the margin must be maximized without any data points violating the hyperplane.



Below is a table that summarizes the differences between hyperplane, marginal plane, soft margin and hard margin in SVM:

| Feature | Hyperplane | Marginal plane | Soft margin | Hard margin |
|---|---|---|---|---|
| Definition | A dividing line that separates two classes of data points. | A parallel hyperplane that is closest to the support vectors. | A SVM that allows some data points to be misclassified. | A SVM that does not allow any data points to be misclassified. |
| Margin | The distance between the hyperplane and the closest data points. | The distance between the hyperplane and the marginal plane. | The number of misclassified data points. | Zero misclassified data points. |

### Q6. SVM Implementation through Iris dataset.

```python
## importing necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
print(dataset.DESCR)
```

```python
dataset.feature_names
```

```python
dataset.target_names
```

```python
X=pd.DataFrame(data=dataset.data,columns=dataset.feature_names)
y=dataset.target
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test =train_test_split(X, y, random_state=42, test_size=0.25)
```

```python
from sklearn.svm import SVC
svc=SVC()
```

```python
svc.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={"kernel":['linear','rbf','poly','sigmoid'],
            "C":[0.1,1,10,100,1000],
            "gamma":[1,0.1,0.01,0.001,0.001],            
}
```

```python
grid=GridSearchCV(svc,param_grid=parameters,verbose=1)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
cm=confusion_matrix(y_test,y_pred)
disp=ConfusionMatrixDisplay(cm,display_labels=svc.classes_)
disp.plot()
plt.show()
```
